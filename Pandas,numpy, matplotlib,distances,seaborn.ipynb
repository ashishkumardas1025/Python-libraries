```python
# euclidean distance
url= df.drop(labels="Country or region", axis=1).values

url
```

```python
url.size
```

```python
np.size(url,0)
```

```python
eucl= np.zeros([np.size(url,0), np.size(url,0)])
eucl
```

```python
for i in range(0,np.size(url,0)):
    for j in range(0,np.size(url,0)):
        eucl[i,j]= sqrt(sum(pow(a-b,2) for a, b in zip(url[i,:],url[j,:])))
        
        
plt.imshow(eucl)
```

```python
manht=np.zeros([np.size(url,0),np.size(url,0)])
for i in range(0,np.size(url,0)):
    for j in range(0,np.size(url,0)):
        manht[i,j]=sum(abs(a-b) for a,b in zip(url[i,:],url[j,:]))
plt.imshow(manht)
```

```python
minkow= np.zeros([np.size(url,0),np.size(url,0)])
root=4
root_value= 1/float(root)
for i in range(0,np.size(url,0)):
    for j in range(0,np.size(url,0)):
        minkow[i,j]= round(Decimal(sum(pow(abs(a-b),root_value) for a,b in zip(url[i,:], url[j,:]))))
        
plt.imshow(minkow)
```

```python
#cosine similarity in python
#here we check to samples are similar or not
cosinesimilarity= np.zeros([np.size(url,0),np.size(url,0)])
for i in range(0,np.size(url,0)):
    for j in range(0, np.size(url,0)):
        numerator=sum(a*b for a,b in zip(url[i,:],url[j,:]))
        denominator=round(sqrt(sum([a*a for a in url[i,:]])),3)*round(sqrt(sum([a*a for a in url[j,:]])),3)
        cosinesimilarity[i,j]= round(numerator/float(denominator),3)
    
plt.imshow(cosinesimilarity)
```
